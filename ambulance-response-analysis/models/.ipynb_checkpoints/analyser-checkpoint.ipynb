```python
import re
import pandas as pd
import warnings
warnings.filterwarnings("ignore", message=".*kurtosistest only valid for n>=20.*")

def parse_cares_summary(path, year):
    rows = []
    current_section = None
    
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            # Section header
            if re.match(r".*,N=\d+", line) and not line.startswith(" "):
                current_section = line.split(",")[0]
                continue

            # Sub-row with data
            if current_section and "," in line:
                parts = line.split(",", 1)
                label = parts[0].strip()
                value = parts[1].strip()

                # Extract count and percent
                match = re.match(r"([\d,]+)\s*\(([\d\.]+)\)", value)
                if match:
                    count = int(match.group(1).replace(",", ""))
                    percent = float(match.group(2))
                else:
                    # sometimes only a count exists
                    count = int(re.sub(r"[^\d]", "", value)) if value else None
                    percent = None

                rows.append({
                    "year": year,
                    "section": current_section,
                    "metric": label,
                    "count": count,
                    "percent": percent
                })

    return pd.DataFrame(rows)

# Example: parse your 2013 file
df2013 = parse_cares_summary("../data/processed/Cares_2013_Summary_Data.csv", 2013)
print(df2013.head(15))
```

```python
all_years = []
for year in range(2013, 2025):
    path = f"../data/processed/Cares_{year}_Summary_Data.csv"
    try:
        df_year = parse_cares_summary(path, year)
        all_years.append(df_year)
    except FileNotFoundError:
        print(f"Missing file: {path}")

df = pd.concat(all_years, ignore_index=True)

# Save clean table
df.to_csv("../data/processed/cares_summary_clean.csv", index=False)
```

```python
# models/trend_and_glm.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import glm, ols
from statsmodels.genmod.families import Binomial

sns.set(style="whitegrid")

# ---------- LOAD your tidy per-year table ----------
# If you already have a dataframe 'df' with columns: year, surv_count, totalN, surv_pct
# Example small loader if you exported surv_pct as shown:
data = {
    "year":[2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024],
    "surv_pct":[10.1,10.8,10.6,10.8,10.4,10.4,10.5,9.0,9.1,9.3,10.2,10.5]
}
df = pd.DataFrame(data)

# If you have counts, load them instead of dummy:
# df = pd.read_csv("../reports/parsed_summary_overview.csv")
# Ensure df has columns ['year','surv_count','totalN']
# df['surv_pct'] = df['surv_count'] / df['totalN'] * 100

# ---------- Visualize raw trend ----------
plt.figure(figsize=(10,5))
plt.plot(df['year'], df['surv_pct'], marker='o', linewidth=2)
plt.title("Overall Survival to Hospital Discharge (%) by Year")
plt.xlabel("Year")
plt.ylabel("Survival to Discharge (%)")
plt.axvline(2020, color='red', linestyle='--', label='2020 (breakpoint)')
plt.legend()
plt.tight_layout()
plt.savefig("../reports/figures/survival_trend.png", dpi=150)
plt.show()

# ---------- Smooth trend (rolling + lowess) ----------
df['surv_ma3'] = df['surv_pct'].rolling(window=3, center=True, min_periods=1).mean()

from statsmodels.nonparametric.smoothers_lowess import lowess
lowess_sm = lowess(df['surv_pct'], df['year'], frac=0.4)
plt.figure(figsize=(10,5))
plt.plot(df['year'], df['surv_pct'], 'o', label='Observed')
plt.plot(df['year'], df['surv_ma3'], '-', label='3-yr MA')
plt.plot(lowess_sm[:,0], lowess_sm[:,1], '-', label='LOWESS')
plt.title("Smoothed Survival Trend")
plt.legend()
plt.tight_layout()
plt.show()

# ---------- 1) Simple linear trend (OLS) on percent ----------
ols_res = ols("surv_pct ~ year", data=df).fit()
print("\nOLS linear trend:")
print(ols_res.summary())

# ---------- 2) Piecewise (segmented) linear model with breakpoint at 2020 ----------
df['post2020'] = (df['year'] >= 2020).astype(int)
df['year_centered'] = df['year'] - df['year'].mean()
df['year_post_slope'] = df['year_centered'] * df['post2020']

seg_model = ols("surv_pct ~ year_centered + post2020 + year_post_slope", data=df).fit()
print("\nSegmented regression (break 2020) results:")
print(seg_model.summary())

# Interpretation hint:
# - coefficient for 'post2020' captures level shift at 2020,
# - 'year_post_slope' captures change in slope after 2020.

# ---------- 3) Binomial GLM (preferred) if you have counts ----------
# Use this block only if surv_count and totalN exist in df
if {'surv_count','totalN'}.issubset(df.columns):
    df['surv_prop'] = df['surv_count'] / df['totalN']
    predictors = ['year_centered','post2020','year_post_slope']  # extend with other predictors
    X = sm.add_constant(df[predictors])
    glm_binom = sm.GLM(df['surv_prop'], X, family=Binomial(), freq_weights=df['totalN'])
    res_binom = glm_binom.fit()
    print("\nBinomial GLM results:")
    print(res_binom.summary())
else:
    print("\nNo per-year counts (surv_count/totalN) found - skipping binomial GLM. If you have counts, include them in df.")

# ---------- 4) Next: multivariable model (example) ----------
# Once you parse other monthly/yearly predictors like bystander_pct, aed_pct, witnessed_pct:
# Formula: surv_prop ~ year_centered + bystander_prop + aed_prop + witnessed_prop + shockable_prop
# Importance: use proportions and GLM Binomial with freq_weights=totalN

```

```python
from lifelines import KaplanMeierFitter

# ---------- Prepare data ----------
# Using counts if you have them, else using percentages
if {'surv_count','totalN'}.issubset(df.columns):
    df['fail_count'] = df['totalN'] - df['surv_count']
    # Kaplan-Meier requires per-individual duration & event: we simulate per-year "time" as year index
    km_data = []
    for _, row in df.iterrows():
        km_data.extend([(row['year'], 1)]*int(row['surv_count']))  # event=1 for survived
        km_data.extend([(row['year'], 0)]*int(row['fail_count']))  # event=0 for not survived
    km_df = pd.DataFrame(km_data, columns=['year','event'])
else:
    # fallback: use proportions to create approximate survival curve
    km_df = pd.DataFrame({
        'year': df['year'],
        'event': df['surv_pct']/100
    })

# ---------- Fit Kaplan-Meier ----------
kmf = KaplanMeierFitter()
# Using year as timeline, event as observed survival
if 'surv_count' in df.columns:
    kmf.fit(durations=km_df['year'], event_observed=km_df['event'])
else:
    kmf.fit(durations=df['year'], event_observed=df['surv_pct']/100)  # approximation

# ---------- Plot survival curve ----------
plt.figure(figsize=(10,6))
kmf.plot_survival_function(ci_show=True)
plt.title("Kaplan-Meier Style Survival Curve (Annual Data)")
plt.xlabel("Year")
plt.ylabel("Survival Probability to Hospital Discharge")
plt.axvline(2020, color='red', linestyle='--', label='2020 breakpoint')
plt.legend()
plt.tight_layout()
plt.show()
```
